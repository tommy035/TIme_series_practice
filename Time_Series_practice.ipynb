```python
column = ['Consumption', 'Solar', 'Wind']
df[column]['2017-12-1':'2017-12-7'].round(1)
```

```python
df_weekly_roll  = df[column].rolling(7, center=True).mean()
df_weekly_roll['2017-12-1':'2017-12-7'].round(1)
```

```python
df_weekly_resampling = df[column].resample('7d').mean()
df_weekly_resampling.tail()
```

```python
df_resampling = df[column].resample('1w').mean()
df_resampling.plot(grid=True)
plt.ylabel('GWh')
plt.title('Power')
plt.show()
```

```python
start = '2017-2'
end = '2017-10'
df['Solar'].loc[start:end].plot(marker='o', label='Daily', markersize=2, alpha=0.7)
df['Solar'].loc[start:end].resample('15d').mean().plot(marker='o', c='blue', markersize=8, label='Mean Sampling')
df['Solar'].loc[start:end].rolling(15, center=True).mean().plot(marker='o', markersize=4, label='Mean Rolling')
plt.ylabel('Power GWh')
plt.title('Solar')
plt.legend()
plt.grid()
plt.show()
```

```python
columns = ['Solar', 'Wind']
yearly = df[columns].resample('y').sum()

yearly.plot(kind='bar', figsize=(10,4), rot=20)
plt.grid()
plt.show()
```

```python
def x_label(datetime):
    lb = str(datetime.year) #+ '-'+str(datetime.month)
    return lb

ax = yearly.plot(kind='bar', figsize=(10,4), rot=20)
ax.set_xticklabels(map(lambda x: x_label(x), yearly.index))
ax.set(xlabel='Year', ylabel='Power GWh')
ax.set(title='Yearly Power 2006-2018')
plt.grid()
plt.show()
```
